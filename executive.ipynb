```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import ast
import sys
import os

sys.path.append('..')

from src.utils import load_csv, load_json
```

```python
metrics = load_csv('../results/metrics.csv')
outputs = load_json('../results/outputs.json')

print("Benchmark Results Summary")
print("=" * 50)
print(f"Overall Accuracy: {metrics['overall_accuracy'].iloc[0]:.2%}")
print(f"Total Tasks: {metrics['total_tasks'].iloc[0]}")
print(f"Correct Tasks: {metrics['correct_tasks'].iloc[0]}")
```

```python
type_acc = ast.literal_eval(metrics['accuracy_by_task_type'].iloc[0])
type_df = pd.DataFrame(list(type_acc.items()), columns=['Task Type', 'Accuracy'])

plt.figure(figsize=(10, 6))
plt.bar(type_df['Task Type'], type_df['Accuracy'])
plt.xlabel('Task Type')
plt.ylabel('Accuracy')
plt.title('Accuracy by Task Type')
plt.ylim(0, 1)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

for task_type, acc in type_acc.items():
    print(f"{task_type}: {acc:.2%}")
```

```python
diff_acc = ast.literal_eval(metrics['accuracy_by_difficulty'].iloc[0])
diff_df = pd.DataFrame(list(diff_acc.items()), columns=['Difficulty', 'Accuracy'])

plt.figure(figsize=(8, 6))
plt.bar(diff_df['Difficulty'], diff_df['Accuracy'])
plt.xlabel('Difficulty')
plt.ylabel('Accuracy')
plt.title('Accuracy by Difficulty')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()

for diff, acc in diff_acc.items():
    print(f"{diff}: {acc:.2%}")
```

```python
trap_acc = ast.literal_eval(metrics['accuracy_by_trap_type'].iloc[0])
trap_df = pd.DataFrame(list(trap_acc.items()), columns=['Trap Type', 'Accuracy'])

plt.figure(figsize=(10, 6))
plt.bar(trap_df['Trap Type'], trap_df['Accuracy'])
plt.xlabel('Trap Type')
plt.ylabel('Accuracy')
plt.title('Accuracy by Trap Type')
plt.ylim(0, 1)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

for trap, acc in trap_acc.items():
    print(f"{trap}: {acc:.2%}")
```

```python
df_outputs = pd.DataFrame(outputs)
incorrect = df_outputs[df_outputs['prediction'].astype(str) != df_outputs['ground_truth'].astype(str)]

print(f"Total Incorrect: {len(incorrect)}")

for _, row in incorrect.head(5).iterrows():
    print("\n---")
    print(f"Task Type: {row['task_type']}")
    print(f"Difficulty: {row['difficulty']}")
    print(f"Trap: {row['trap_type']}")
    print(f"Question: {row['question']}")
    print(f"Truth: {row['ground_truth']}")
    print(f"Prediction: {row['prediction']}")
```

```python
print("\nKey Insights")
print("=" * 50)

for trap, acc in trap_acc.items():
    if acc < 0.5:
        print(f"Low performance on {trap}: {acc:.2%}")

for diff, acc in diff_acc.items():
    if acc < 0.6:
        print(f"Drop in {diff}: {acc:.2%}")

best_type = max(type_acc, key=type_acc.get)
worst_type = min(type_acc, key=type_acc.get)

print(f"\nBest Task Type: {best_type} ({type_acc[best_type]:.2%})")
print(f"Worst Task Type: {worst_type} ({type_acc[worst_type]:.2%})")
```
